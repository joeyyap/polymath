```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())
```

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.penanghawkers.my/explore/?category=hawker&sort=a-z"

driver.get(url)
```

```python
import pandas as pd
import time


headers = ['Name', 'Contact', 'Link']

namelist = []
contactlist = []
linklist = []
```

```python

clicknext = True

while clicknext:
    
    soup = bs(driver.page_source, 'html.parser')
#     time.sleep(5)
    
    for item in soup.find_all('div', attrs = {'class':'lf-item lf-item-default'}):
        name = item.find('h4', attrs = {'class':'listing-preview-title'}).text.strip()
        link = item.find('a')['href']
        contact = item.find('li').text.strip()

        namelist.append(name)
        contactlist.append(contact)
        linklist.append(link)
        
        print(name)

#     last_button = soup.find('div', attrs = {'class':'load-next-batch load-batch-icon batch-unavailable'})
    
#     print(last_button)
    
#     if type(last_button) != type(None):
#         clicknext=False
        
#     else:
    driver.find_element_by_class_name('load-next-batch').click()
```

```python
data = list(zip(namelist, contactlist, linklist))

df = pd.DataFrame(data,columns=headers)
df
```

```python
driver.quit()
```
